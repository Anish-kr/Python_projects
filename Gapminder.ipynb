### How has employment in U.S. varied over time?

```python
emp_US=employment_rate.loc['United States']
emp_US_df = pd.DataFrame(emp_US)
```

```python
# finding correlation

def correlation( x,y):
    std_x = (x-x.mean())/(x.std(ddof = 0))
    std_y = (y-y.mean())/(y.std(ddof=0))
    
    return (std_x * std_y).mean()

print emp_US.values
year_for_US=employment_rate.columns.values.astype(str).astype(int)
```

```python
correlation(emp_US.values, year_for_US)
```
